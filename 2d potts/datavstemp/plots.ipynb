```python
def parabola(x, a, b, c):
    return a * x**2 + b * x + c

def k_max_and_fit(data, errors, k):
    
    k_max_elements_with_index = [(data[i], i) for i in range(len(data))]
    k_max_elements_with_index.sort(reverse=True, key=lambda x: x[0])
    k_max_elements_with_index = k_max_elements_with_index[:k]
    
    # Estrarre le posizioni corrispondenti ai k elementi massimi
    x_data = np.array([t[x[1]] for x in k_max_elements_with_index])
    y_data = np.array([x[0] for x in k_max_elements_with_index])
    dy_data = np.array([errors[x[1]] for x in k_max_elements_with_index])
    
    # Fit i dati con un modello di parabola
    params, cov = curve_fit(parabola, x_data, y_data,sigma=dy_data)
    
    # Calcolare gli errori sui parametri
    err = np.sqrt(np.diag(cov))
    
    # Calcolare il chi-quadrato
    residuals = y_data - parabola(x_data, *params)
    chisq = np.sum((residuals / dy_data)**2)
    
    beta=-params[1]/(2*params[0])
    max=parabola(beta,*params)
    return beta,max
```

```python
def beta_and_max(start,stop,data,errors):
    betas_maxs=[]
    betas=[]
    maxs=[]
    for i in range (start,stop):
        betas_maxs.append(k_max_and_fit(data,errors,i))
    for values in betas_maxs:
        betas.append(values[0])
        maxs.append(values[1])
    beta=mean(betas)
    max=mean(maxs)
    d_beta=stdev(betas)
    d_max=stdev(maxs)
    return beta,d_beta,max,d_max
```

```python
betas=[]
dbetas=[]
xmaxs=[]
dxmaxs=[]
for i in range (0, len(m)):
    beta,dbeta,xm,dxm=beta_and_max(3,8,x[i],dx[i])
    betas.append(beta)
    dbetas.append(dbeta)
    xmaxs.append(xm)
    dxmaxs.append(dxm)
```

```python
def scaling_beta(x, a, b, c):
    return a * x**(-1/b) + c
```

```python
params, cov = curve_fit(scaling_beta, L,betas,sigma=dbetas)

# Calcolare gli errori sui parametri
errors = np.sqrt(np.diag(cov))

# Calcolare il chi-quadrato
residuals = np.array(betas) - scaling_beta(np.array(L), *params)
chisq = ((residuals/dbetas)**2).sum()

print(chisq)
print(params)
print(errors)
```

```python
b_c=params[2]
db_c=errors[2]
nu=params[1]
dnu=errors[1]
```

```python
n=np.linspace(min(L)-1,max(L)+1,10000)
pylab.figure(figsize=(13, 11))
pylab.title(r'fit $\nu$')
pylab.rc('font', size = 15)
pylab.xlabel('L', labelpad = 25)
pylab.ylabel(r'$\beta$')
pylab.grid(color = 'silver')
pylab.errorbar(L, betas, dbetas, linestyle = '', color = 'black', marker = '.')
pylab.plot(n, scaling_beta(n,*params), color = 'blue', label = 'retta di fit')

pylab.show()
```

```python
def scaling_x(x, a, b, c):
    return a * x**b + c
```

```python
params, cov = curve_fit(scaling_x, L,xmaxs,sigma=dxmaxs)

# Calcolare gli errori sui parametri
errors = np.sqrt(np.diag(cov))

# Calcolare il chi-quadrato
residuals = np.array(xmaxs) - scaling_x(np.array(L), *params)
chisq = ((residuals/dxmaxs)**2).sum()

print(chisq)
print(params)
print(errors)
```

```python
n=np.linspace(min(L)-1,max(L)+1,10000)
pylab.figure(figsize=(13, 11))
pylab.title(r'fit $\gamma/\nu$')
pylab.rc('font', size = 15)
pylab.xlabel('L', labelpad = 25)
pylab.ylabel(r'$\chi_{max}$')
pylab.grid(color = 'silver')
pylab.errorbar(L, xmaxs, dxmaxs, linestyle = '', color = 'black', marker = '.')
pylab.plot(n, scaling_x(n,*params), color = 'blue', label = 'retta di fit')

pylab.show()
```

```python
gamma=nu*params[1]
dgamma=gamma*np.sqrt((dnu/nu)**2+(errors[1]/params[1])**2)
ni=2-params[1]
dni=errors[1]
alpha=2-2*nu
dalpha=2*dnu
b=(2-gamma-alpha)/2
db=np.sqrt(dgamma**2+dalpha**2)/4
delta=gamma/b+1
d_delta=delta*np.sqrt((dgamma/gamma)**2+(db/b)**2)
```
